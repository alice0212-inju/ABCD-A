```python
# 1. YOLOv8 설치
!pip install ultralytics --upgrade -q

# 2. 라이브러리 불러오기
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

# 3. 모델 불러오기 (n은 가장 가벼운 모델, s/m/l/x로 성능/속도 선택 가능)
model = YOLO("yolov8n.pt")  # 또는 yolov8s.pt

# 4. 이미지 업로드
uploaded = files.upload()

# 업로드된 이미지 리스트
image_paths = list(uploaded.keys())

# 5. 이미지별로 고양이 탐지
for path in image_paths:
    print(f"📷 이미지: {path}")

    # 이미지 로드
    img = cv2.imread(path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 모델 추론
    results = model(img_rgb)[0]

    # "cat" 클래스 ID는 COCO 데이터셋 기준 15번입니다.
    # 결과에서 클래스 ID가 15인 것만 필터링
    cat_boxes = [box for box in results.boxes if int(box.cls) == 15]
    cat_count = len(cat_boxes)

    # 결과 시각화
    annotated_img = results.plot()
    plt.imshow(annotated_img)
    plt.axis('off')
    plt.title(f"Detected Cats: {cat_count}")
    plt.show()
```
