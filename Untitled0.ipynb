```python
# 1. YOLOv8 ÏÑ§Ïπò
!pip install ultralytics --upgrade -q

# 2. ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files

# 3. Î™®Îç∏ Î∂àÎü¨Ïò§Í∏∞ (nÏùÄ Í∞ÄÏû• Í∞ÄÎ≤ºÏö¥ Î™®Îç∏, s/m/l/xÎ°ú ÏÑ±Îä•/ÏÜçÎèÑ ÏÑ†ÌÉù Í∞ÄÎä•)
model = YOLO("yolov8n.pt")  # ÎòêÎäî yolov8s.pt

# 4. Ïù¥ÎØ∏ÏßÄ ÏóÖÎ°úÎìú
uploaded = files.upload()

# ÏóÖÎ°úÎìúÎêú Ïù¥ÎØ∏ÏßÄ Î¶¨Ïä§Ìä∏
image_paths = list(uploaded.keys())

# 5. Ïù¥ÎØ∏ÏßÄÎ≥ÑÎ°ú Í≥†ÏñëÏù¥ ÌÉêÏßÄ
for path in image_paths:
    print(f"üì∑ Ïù¥ÎØ∏ÏßÄ: {path}")

    # Ïù¥ÎØ∏ÏßÄ Î°úÎìú
    img = cv2.imread(path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Î™®Îç∏ Ï∂îÎ°†
    results = model(img_rgb)[0]

    # "cat" ÌÅ¥ÎûòÏä§ IDÎäî COCO Îç∞Ïù¥ÌÑ∞ÏÖã Í∏∞Ï§Ä 15Î≤àÏûÖÎãàÎã§.
    # Í≤∞Í≥ºÏóêÏÑú ÌÅ¥ÎûòÏä§ IDÍ∞Ä 15Ïù∏ Í≤ÉÎßå ÌïÑÌÑ∞ÎßÅ
    cat_boxes = [box for box in results.boxes if int(box.cls) == 15]
    cat_count = len(cat_boxes)

    # Í≤∞Í≥º ÏãúÍ∞ÅÌôî
    annotated_img = results.plot()
    plt.imshow(annotated_img)
    plt.axis('off')
    plt.title(f"Detected Cats: {cat_count}")
    plt.show()
```
